```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# BÃ¡sico
xvals = np.arange(0, 10, 0.1)
plt.plot(xvals)
plt.xlabel('$x$')
plt.ylabel(r'$sin(x)$')
```

```python
idx = 0
for marker in ('-', '--', ':', '-.'):
    plt.plot(xvals, np.sin(xvals)+idx, ''.join(('k', marker)))
    idx += 1
```

```python
# Marcadores
plt.plot(xvals, np.sin(xvals), 'r.', markersize=1, markevery=3)
plt.xlabel('$x$')
plt.ylabel(r'$sin(x)$')
```

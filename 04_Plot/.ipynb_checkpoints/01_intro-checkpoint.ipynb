```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración general de los plots (revisar documentación)
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Básico
xvals = np.arange(0, 10, 0.1)
plt.plot(xvals)
```

```python
# Ejes
xvals = np.arange(0, 10, 0.1)
plt.plot(xvals, np.sin(xvals))
```

```python
# xdas
plt.plot(xvals, np.sin(xvals))
plt.plot(xvals, np.cos(xvals))
plt.xlabel('$x$')
plt.ylabel(r'$y$')
```

```python
# label en el gráfico
plt.plot(xvals, np.sin(xvals), label = r'$sin(x)$')
plt.plot(xvals, np.cos(xvals), label = r'$cos(x)$')
plt.plot(xvals, np.tan(xvals), label = r'$tan(x)$')
plt.xlabel('$x$')
plt.ylabel(r'$y$')
plt.legend()
```

```python
# label en el gráfico y colores
plt.plot(xvals, np.sin(xvals), label = '$sin(x)$', color = 'r')
plt.plot(xvals, np.cos(xvals), label = '$cos(x)$', color = 'b')
plt.plot(xvals, np.tan(xvals), label = '$tan(x)$', color = 'k')
plt.xlabel('$x$')
plt.ylabel(r'$y$')
plt.legend()
```

```python
# label en el gráfico, colores y tipo de línea
plt.plot(xvals, np.sin(xvals), label = '$sin(x)$', color = 'r', marker = '*')
plt.plot(xvals, np.cos(xvals), label = '$cos(x)$', color = 'b', marker = '*')
plt.xlabel('$x$')
plt.ylabel(r'$y$')
plt.legend()
```

```python
# Con el 3er argumento
plt.plot(xvals, np.sin(xvals), 'r--', label = '$sin(x)$')
plt.plot(xvals, np.cos(xvals), 'k*', label = '$cos(x)$')
plt.xlabel('$x$')
plt.ylabel(r'$y$')
plt.legend()
```

```python
# límites
plt.plot(xvals, np.sin(xvals), 'r--', label = '$sin(x)$')
plt.plot(xvals, np.cos(xvals), 'k*', label = '$cos(x)$')
plt.xlim(0,2)
plt.ylim(0,2)
plt.xlabel('$x$')
plt.ylabel(r'$y$')
plt.legend()
```

```python
rands = np.random.normal(size=1000)
print(rands)
plt.xlabel('Iteration #')
plt.ylabel('Random value')
```

```python
# Campo Gausiano
from scipy.ndimage.filters import gaussian_filter

rands2d = gaussian_filter(np.random.normal(size=(512,512)), sigma=10)
print(rands2d)

plt.imshow(rands2d, cmap='inferno')
plt.colorbar()
```
